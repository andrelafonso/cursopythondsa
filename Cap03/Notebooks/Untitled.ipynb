```python
import pandas as pd
from pandas_datareader import data as web
import plotly.graph_objects as go

df = pd.DataFrame()

acao = 'VVAR3.SA'

df = web.DataReader(acao, data_source='yahoo', start='01-01-2019')

df.head()

trace1 = {
    'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': acao,
    'showlegend': True
}

data = [trace1]
layout = go.Layout()

fig = go.Figure(data=data, layout=layout)
fig.show()
```
